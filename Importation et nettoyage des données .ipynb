# Nettoyage des données 

```python
#importation des donnees
df = pd.read_csv('data_nlp.csv')

# Remplacer les caractères de nouvelle ligne par des espaces vides
df['text'] = df['text'].replace('\n', '', regex=True)
```

### Fonctions pour le nottoyage des données 

```python
### NETTOYAGE REGEX
def nettoyage_regex(text):
    text = re.sub(r'\d+', '', text)  # Enlève les chiffres
    text = re.sub(r'\b\w\b', '', text)  # Enleve les lettres isolées
    return text

### TOKENISATION
def tokenisation(text):
    tokenizer = nltk.tokenize.WordPunctTokenizer()
    tweet_tokens = [tokenizer.tokenize(text)]
    return tweet_tokens[0]

### STOP WORDS
stop_words = stopwords.words('french')
new_stopwords = ["avoir", "être", "article", "loi", "cour", "appel", "article", "code",
                "chambre", "civile", "cassation", "procédure", "x", "y", "z", "mme", "mlle",
                "janvier", "février", "mars", "avril", "mai", "juin", "juillet", "août",
                "septembre", "octobre", "novembre", "décembre", "lundi", "mardi",
                "mercredi", "jeudi", "vendredi", "samedi", "dimanche","a", "e", "b",
                "deuxième", "rendu" ,"arrêt" ,"suivant" ,"moyen" ,"unique", "selon",
                "demande" ,"avis", "séance" ,"juridiction", "tribunal" ,"grande" ,"instance",
                "nom" ,"peuple" ,"francais" ,"deuxieme" , "premier","premiere", "première", "troisième" ,
                "pourvoi", "pris",
                "formé", "criminelle", "audience", "publique", "paris", "tenue", "palais", "justice",
                "dont", "siège", "f", "r"]
stop_words.extend(new_stopwords)
stop_words = set(stop_words)
def stop_words_function(text):
    return [w.lower() for w in text if not w.lower() in stop_words]

## SUPPRESSION PONCTUATION
def suppr_ponct(text):
    return [token for token in text if token.isalnum()]

## SUPPRESSION ACCENTS
def suppr_accent(text):
    return [unidecode(token) for token in text if token]


### LEMMATISATION
nlp = spacy.load('fr_core_news_md')
def lemmatisation(text):
    lem = []
    for i in range(len(text)):
        doc = nlp(text[i])
        for token in doc:
            lem.append(token.lemma_)
    return lem

def nettoyage(text):
    text = nettoyage_regex(text)
    text = tokenisation(text)
    text = suppr_ponct(text)
    #text = lemmatisation(text)
    text = stop_words_function(text)
    text = suppr_accent(text)
    return text

def nettoyage_phrase(text):
    text = nettoyage_regex(text)
    text = tokenisation(text)
    return text
```

### Construction et exportation d'un premier jeu de données nettoyé

Nous allons exporter le jeu de données nettoyé de 2 façons. Nous auront besoin de ces deux formats de données pour la suite de l'analyse. Dans un premier temps, on exporte les textes nettoyés grâce à la fcontion nettoyage qui comporte :

    - Un nettoyage Regex
    - Une tokenisation
    - Suppression de la ponctuation
    - Suppression des stopwords
    - Supression des accents
Le jeu de données en sortie comporte donc pour chaque ligne du tableau une liste des mots "utiles" du texte. On l'exporte afin de l'utiliser dans la suite de l'analyse.

```python
# Création d'une liste vide pour stocker le texte nettoyé
cleaned_text = []
# On boucle sur tous les textes 
for text in df['text']:
    # Application de la fonction nettoyage
    cleaned = nettoyage(text)
    # Ajout du texte nettoyé à la liste 
    cleaned_text.append(cleaned)
    
cleaned_text = [' '.join(words) for words in cleaned_text]
cleaned_text_df = pd.DataFrame().assign(text=cleaned_text)
cleaned_text_df.to_csv("cleaned_text2.csv")
```

### Contruction et exportation d'un second jeu de données nettoyé

Dans un second temps, on exporte un jeu de données qui contient les phrases entières (on ne retire ni la ponctuation ni les stopwords). Ce jeu de données nous permettra par la suite d'étudier notamment le contexte d'apparition des mots ou encore le graphique des dépendances en utilisant des phrases complètes. De même que pour le précédent, on exporte le jeu de données pour pouvoir l'utiliser dans la suite de l'analyse.

```python
# Création d'une liste vide pour stocker le texte nettoyé
cleaned_text = []
# On boucle sur tous les textes 
for text in df['text']:
    # Application de la fonction nettoyage
    cleaned = nettoyage_phrase(text)
    # Ajout du texte nettoyé à la liste 
    cleaned_text.append(cleaned)
    
cleaned_text = [' '.join(words) for words in cleaned_text]
cleaned_text_df = pd.DataFrame().assign(text=cleaned_text)
cleaned_text_df.to_csv("cleaned_phrase2.csv")
```
